## All files that we have for Bondville scattering

```python
import pyaerocom as pya
import numpy as np

scat_name = 'aerosol_light_scattering_coefficient'
rh_name = 'relative_humidity'
```

```python
request = pya.io.EbasSQLRequest(variables=scat_name, station_names='Bondville')
```

```python
files = pya.io.EbasFileIndex().get_file_names(request)
print('Total no. of files: {}\n\n'.format(len(files)))
```

```python
for i, file in enumerate(files):
    data = pya.io.EbasNasaAmesFile(pya.const.EBASMC_DATA_DIR + file)
    start, stop = data.time_stamps[0], data.time_stamps[-1],
    print("FILE {} ({}-{})\n----------------------------------".format(i+1, start, stop))
    print('Filename:{}\nStart: {}, Stop:{}\nNumber of datapoints: {}'.format(data.file_name, 
                                                                             start, stop,
                                                                             data.shape[0]))
    for colnum, info in enumerate(data.var_defs):
        if info.name == scat_name and 500 <= info.get_wavelength_nm() <= 600 and info.statistics=='arithmetic mean': # valid column
            numnotnan = np.sum(~np.isnan(data.data[:, colnum]))
            print('\nFound valid data column')
            print(repr(info))
            print('Number of datapoints that are not NaN: {}'.format(numnotnan))
        if info.name == rh_name:
            numnotnan = np.sum(~np.isnan(data.data[:, colnum]))
            print('\nFound RH column')
            print(repr(info))
            print('Number of datapoints that are not NaN: {}'.format(numnotnan))
            
        
    print('\n\n')
            
```
